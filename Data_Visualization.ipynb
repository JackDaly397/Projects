```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes = True)
```

```python
data = pd.read_csv('/syn_crab_data (2).csv')
```

```python
data.head(10)
```

```python
sns.set(rc = {'figure.figsize':(20,8)})
```

```python
data.plot(kind='scatter', x = 'Length', y = 'Diameter')
plt.title('Crab Length vs Diameter')
```

```python
data.plot(kind='scatter', x = 'Diameter', y = 'Weight')
plt.title('Crab Diameter vs Weight')
```

```python
joined =sns.jointplot(x = 'Length', y = 'Weight', data = data, size = 10)
joined.fig.subplots_adjust(top=.9)
joined.fig.suptitle('Jointplot showing crab length vs crab weight')
plt.show()
```

```python
rp = sns.pairplot(data[['Length','Diameter','Weight','Sex','Age']].sample(8000), hue='Sex', diag_kind="hist", palette='Paired')
rp.fig.subplots_adjust(top=0.9) # adjust the Figure in rp

rp.fig.suptitle('Pairplot showing bivariate relations')
plt.show()
```

```python
sns.histplot(x='Age', data=data)
plt.title('Distribution of Crab Ages')
plt.show()
```

```python
sns.countplot(x='Age', hue='Sex', palette='Set3', data=data, hue_order=['M', 'F','I'])
plt.title('Distribution of Crab Age split by Sex')
plt.show()
```

```python
sns.boxplot(x='Age', y='Weight', palette='husl', data=data).set(title='Boxplot showing Crab Age vs Weight')
```

```python
sns.boxplot(x='Sex', y='Weight', data=data, order = ['M','F','I'])
plt.title('Frequency of Crab Weights by Sex')
plt.show()
```

```python
sns.boxplot(x='Sex', y='Age', data=data, order = ['M','F','I'])
plt.title('Frequency of Crab Ages by Sex')
plt.show()
```

```python
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='crest')
plt.title('Correlation Between Features')
plt.show(),
```
